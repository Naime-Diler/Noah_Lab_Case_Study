```python
# Definition der Kriterien und Gewichtungen
kriterien_gewichtungen = {
    "Größe der Praxis": 0.3,
    "Spezialisierung und Bedarf": 0.3,
    "Technische Affinität": 0.2,
    "Erreichbarkeit und Kommunikationsbereitschaft": 0.2
}

# Bewertung der Praxen
praxen = {
    "Dr. med. Torsten Morschheuser": {
        "Größe der Praxis": 3,
        "Spezialisierung und Bedarf": 4,
        "Technische Affinität": 4,
        "Erreichbarkeit und Kommunikationsbereitschaft": 4
    },
    "Prof. Dr. med. Kai Mortensen": {
        "Größe der Praxis": 5,
        "Spezialisierung und Bedarf": 5,
        "Technische Affinität": 5,
        "Erreichbarkeit und Kommunikationsbereitschaft": 4
    }
}

# Funktion zur Berechnung der Gesamtpunktzahl
def gesamtpunktzahl_berechnen(praxis):
    gesamtpunktzahl = 0
    for kriterium, gewicht in kriterien_gewichtungen.items():
        gesamtpunktzahl += praxis[kriterium] * gewicht
    return gesamtpunktzahl

# Priorisierung der Praxen
sortierte_praxen = sorted(praxen.items(), key=lambda x: gesamtpunktzahl_berechnen(x[1]), reverse=True)

# Ausgabe der Priorisierung
print("Priorisierung der Praxen:")
print("{:<40} {:<10}".format("Praxis", "Gesamtpunktzahl"))
for praxis, punktzahl in sortierte_praxen:
    print("{:<40} {:.2f}".format(praxis, gesamtpunktzahl_berechnen(punktzahl)))
```
